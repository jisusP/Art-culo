```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/jisusP/Articulo/main/similarity2.csv"
df = pd.read_csv(url, header=None)

# Preprocesa los datos si es necesario
df = df.drop(df.columns[0], axis=1)
matrix = df.values
lab = ["1 (62)","2 (>250)","3 (>250)","4 (62)","5 (125)","6 (>250)","7 (>250)","8 (125)","9 (250)","10 (>250)","11 (62)","12 (125)"]

df
```

```python
matrix_d = 100 - matrix
plt.figure(figsize=(18,5))
plt.title("CÃ¡ndida Albicans (TARIS)")
d = shc.dendrogram(shc.linkage(matrix, method = "ward"), labels = lab)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



# Crear el heatmap sin mostrar los nÃºmeros
plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=False, cbar=True, square=True, cmap='viridis')
plt.xticks(np.arange(len(lab)) + 0.5, lab, rotation=45, ha='right')
plt.yticks(np.arange(len(lab)) + 0.5, lab, rotation=45, ha='right')
# Configurar el aspecto del heatmap
plt.title("Mapa de calor (TARIS)")
plt.xlabel
plt.ylabel

# Mostrar el heatmap
plt.show()
```
