```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

# Lee el archivo .dat en un DataFrame
df = pd.read_csv('/content/similarity2.csv', header=None)

# Preprocesa los datos si es necesario
df = df.drop(df.columns[0], axis=1)
matrix = df.values
lab = ["6 (62/250)","7 (>250)","8 (>250)","9 (62/250)","10 (125/250)","11 (>250)","12 (>250)","13 (125/>250)","14 (250/>250)","15 (>250)","16 (62/250)","17 (125/>250)"]

df
```

```python

plt.figure(figsize=(18,5))
plt.title("Cándida Albicans (TARIS)")
d = shc.dendrogram(shc.linkage(matrix, method = "ward"), labels = lab)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



# Crear el heatmap sin mostrar los números
plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=False, cbar=True, square=True, cmap='viridis')
plt.xticks(np.arange(len(lab)) + 0.5, lab, rotation=45, ha='right')
plt.yticks(np.arange(len(lab)) + 0.5, lab, rotation=45, ha='right')
# Configurar el aspecto del heatmap
plt.title("Mapa de calor (TARIS)")
plt.xlabel
plt.ylabel

# Mostrar el heatmap
plt.show()
```
